```python
import math

def sigmoid(z):
    return 1 / (1 + math.exp(-z))

x = float(input("Enter x: "))
w = float(input("Enter w: "))
b = float(input("Enter b: "))

z = x * w + b

if z > 0:
    print("Perceptron output:", 1)
else:
    print("Perceptron output:", 0)

sig = sigmoid(z)
print("Sigmoid output:", sig)
```

```python
# saturation caused by weight
x = 1
b = 0
w = 1

for i in range(6):
    z = x * w + b
    print(w, "   |   ", sigmoid(z))
    w = w * 2   # increase weight
```

```python
W1 = [[0.5, -0.3],
      [0.8,  0.2],
      [-0.6, 0.1]]

b1 = [0.1, -0.2, 0.05]
W2 = [0.7, -0.4, 0.9]
b2 = 0.2
x = [1.0, 0.5]
# weight matrix and bias vector
```

```python
hidden_z = []
hidden_a = []

for i in range(3):
    z = W1[i][0]*x[0] + W1[i][1]*x[1] + b1[i]
    a = sigmoid(z)
    hidden_z.append(z)
    hidden_a.append(a)

print("Hidden layer z values:", hidden_z)
print("Hidden layer activations:", hidden_a)
```

```python
z_out = (
    W2[0]*hidden_a[0] +
    W2[1]*hidden_a[1] +
    W2[2]*hidden_a[2] +
    b2
)

output = sigmoid(z_out)

print("Output layer z value:", z_out)
print("Final output:", output)
```
